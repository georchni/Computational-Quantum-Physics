```python
import numpy as np
from matplotlib import pyplot as plt
import random
```

```python
N = 4069
Niter = 1000000

species = np.random.rand((N))
mutated = []
dists = []
mutated_barriers = []
dt = []
tc = 0.01


for _ in range(Niter):
    weakest = np.argmin(species)
    dt.append(np.exp(species[weakest]/tc))
    if len(mutated)>0:
        dists.append(min(abs(weakest-mutated[-1]),abs(weakest+species.size-mutated[-1]),abs(weakest-species.size-mutated[-1])))
    mutated.append(weakest)
    mutated_barriers.append(species[weakest])
    species[weakest] = random.random()
    species[(weakest+1)%species.size] = random.random()
    species[(weakest-1)%species.size] = random.random()
    
    
```

```python
plt.hist(dists,300,log = True)
plt.xscale('log')
plt.xlim(10,100)
plt.ylim(10,10000)
plt.show()
```

```python
plt.hist(mutated_barriers,density=True)
plt.hist(species,density=True)
plt.show()
```
